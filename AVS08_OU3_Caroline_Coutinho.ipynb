## Parte 4: Recuperação da acetona no sistema.
Para o cálculo da recuperação iremos adotar os seguintes parâmetros:

Vazão de alimentação: 50 kg/h

Como a vazão de solvente é metade em massa da corrente de alimentação:

Vazão de solvente: 25 kg/h



Além disso, iremos utilizar a seguinte expressão para o cálculo: 

6.png

Primeiro precisamos encontrar o valor de $E_1$. 

Iremos utilizar as equações de balanço para acetona e o balanço global do sistema: 

**Balanço para a Acetona:** $0.25E_1 + 0.02R_N = 7.5$

**Balanço Global:** $E_1 + R_N = 75$

```python
import numpy as np
```

```python
v_res = np.array([7.5,75])
v_res
```

```python
A_bmg = np.array([[0.25,0.02],
                  [1,1]])
                  
A_bmg
```

```python
# Resolvendo o sistema teremos: 
x_sol_bmg = np.linalg.inv(A_bmg) @ v_res # multiplicando a matriz inversa pela B (matriz dos resultados)
x_sol_bmg # V e L respectivamente
```

Temos: $E_1=26.08$ kg/h e $R_N=48.1$ kg/h

```python
# Utilizando a equação para recuperação teremos:

Rec = (0.25*x_sol_bmg[0] - 0*25)/(0.15*50)
Rec
```

Portanto a recuperação para essas condições será de cerca de 87%.
