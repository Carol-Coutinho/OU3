## Parte 1: Contextualizando a operação unitária





**Sobre o Dióxido de Enxofre:**

O dióxido de enxofre ($SO_2$) é um gás incolor solúvel em água e que é emitido em diversos processos industriais como: produção de ácido sulfúrico, queima de carvão em termoelétricas, produção de fertilizantes, fundição de metais, entre outros. 

Ele pode ser utilizado como conservador e clareador em indústrias químicas, agente selador, entre outras aplicações. No entanto, é importante ressaltar que esse gás deve ser controlado em processos industriais no que tange a sua emissão para a atmosfera. Isso é necessário pois esse gás em altas concentrações na atmosfera pode gerar ácido sulfúrico resultando no fenômeno de chuva ácida. 

As principais características do dióxido de enxofre são mostradas a seguir: 

image.png

Desse modo, a absorção de $SO_2$ presente no ar parece uma alternativa viável para processos que emitem esse gás. Mais do que isso, como é um composto solúvel em água podemos utilizá-la como solvente no processo de absorção. 

**Sobre a operação de absorção:** 

Basicamente, a absorção consiste na transferência de um componente de uma fase gasosa para uma fase líquida no qual o líquido é não-volátil nas condições de operação da coluna. A operação inversa se chama de desabsorção.

Esse mecanismo pode ocorrer com uma força motriz física, no qual o gás é removido devido a diferença de solubilidade em relação ao solvente, ou com uma força motriz química, no qual o gás reage com o solvente e permanece em solução. 

Essa operação é muito utilizada na remoção de contaminantes de gases, remoção de impurezas, recuperação de produtos químicos, entre outros. 

Ao se trabalhar com essa operação unitária é necessario levar em conta o solvente ideal para o processo e o tipo de estrutura da torre (coluna com pratos ou recheadas). 

A imagem a seguir demonstra a operação de abrosção e de desbsorção.

**Sobre o solvente:** 

A escolha do solvente é parte importante da absoração e por isso deve ser escolhido de forma que a separação ocorra de forma eficiente. 

Por isso, devemos atentar para:

- A solubilidade do soluto deve ser elevada para se obter uma maior velocidade de absorção;
- A natureza química do soluto e do solvente devem ser semelhantes para aumentar a solubilidade;
- O solvente deve ter baixa pressão de vapor (baixa volatilidade);
- O solvente deve ter baixa viscosidade para a velocidade de absorção e a transferência de calor serem elevadas e os custos de bombagem serem baixos.
- O solvente deve ser barato e acessível, não tóxico, não inflamável, estável quimicamente e não corrosivo para não encarecer o material de construção do equipamento.



**Sobre o recheio:** 

Outro ponto importante é a escolha do recheio  utilizado na coluna de absoração, que tem como objetivo otimizar o processo de trasnferência de massa devido a sua elevada área por unidade de volume. 

Os recheios podem ser classificados em:

- Randômico
- Estruturados

Sendo que sua construção pode ser feita de diversos materiais como cerâmica, metal e até mesmo polímeros.

Normalmente recheios randômicos de metal são preferíveis devido a maior resistência e boa molhabilidade, porém o custo é mais elevado
comparado aos recheios randômicos de plástico ou cerâmica. 

Recheios de plástico são mais baratos, porém tem uma molhabilidade ruim em pequenos fluxos de líquido.

Recheios cerâmicos tem uma molhabilidade superior, porém, são pouco resistentes
mecanicamente, sendo usados para ambientes corrosivos e de altas temperaturas.

### Obtenção da constante de Henry: 

Para isso iremos utilizar o material ''Compilation of Henry’s law constants (version 4.0) for water as solvent (doi:10.5194/acp-15-4399-2015)''

Nele podemos encontrar os valores de $H^{cp}$ para o $SO_2$ em água.
Através desse valor podemos encontrar a constente de Henry que será utilizada nos futuros cálculos.

O $H^{cp}$ escolhido para esse caso está taxado em amarelo na imagem a seguir:

3.png

```python
H_cp = 1.2e-2 # mol/m³.Pa
```

A partir do $H^{cp}$ podemos calcular o $H^{xp}$ da seguinte forma:

$H^{xp} =  \frac{MM_{H2O}}{\rho_{agua}} \cdot H^{cp}$ 

Si: $[Pa^{-1}]$

```python
H_xp = (M_agua/rho_agua)*H_cp
H_xp # Pa^-1
```

A constante de Henry será encontrada :

$H = \frac{1}{(H_{xp}\cdot P)}$ 

P em Pa (1 atm = 101325 Pa)

```python
H = 1/(H_xp*101325)
H
```

### Traçando o gráfico da reta de equilíbrio e da reta de operação:

Primeiro iremos definir que 95% do $SO_2$ foi absorvido.

yat = 1.0/(20 + 1.0 )

Calculando a razão molar (Ya):

$ Ya = \frac{ya_{eq}}{1-ya_{eq}}$

```python
xa_eq = ya_eq/H
xa_eq
```

### Cálculo do diâmetro da coluna:

Nessa etapa iremos fazer o cálculo do diâmetro da coluna levando em consideração um valor de f_op definido como 0.8. 

Para isso, iremos seguir os seguintes passos:

- Utilizar o gráfico de correlação generalizada de perda de carga. Como temos os dados fluidinâmios dos componentes, conseguimos calcular a abscissa do gráfico e rebatendo na linha de inundação, conseguimos encontrar o Y de inundação equivalente. 

- Usar: $ \phi_G^{inu} = \sqrt(\frac{Y_{cgpc}^{inu} \cdot \rho_G \cdot \rho_L \cdot C}{cf \cdot \psi \cdot \mu^{0.2}})$ (1)

- Usar: $Dc = \sqrt(\frac{4 \cdot m_G}{\epsilon \cdot \pi \cdot \phi_G })  $ (2)

Temos o seginte gráfico de correlação generalizada de perda de carga:

```python
# Calculando a fração mássica com base de cálculo: 100 mol
fm = [100*0.2,100*0.8]
massa_comp = fm[0]*MM_SO2 + fm[1]*MM_ar
comp_mass = [fm[0]*MM_SO2/massa_comp,fm[1]*MM_ar/massa_comp]

comp_mass
```

```python
rho_gas = rho_SO2*comp_mass[0] + rho_ar*comp_mass[1]
```

```python
# Calculando a abscissa:
x = (Ls_mass/F_mass)*((rho_gas/rho_agua)**0.5)
x
```

Rebatendo o valor de x na curva de inundação, encontramos o valor de Y de inundação.

 Agora iremos calcular o fluxo de inundação através da equação (1).

Onde: 

C = 2.994

$\psi = \rho_{H_2O}/ \rho_L$

$μ_L$ = cP (centipoise)

$c_f$ é o fator de empacotamento do recheio

$rho = [kg/m³]$

```python
C = 2.994
psi = 998/rho_agua
```

### Cálculo da vazão mássica máxima de solvente correspondente ao diâmetro encontrado:

Iremos seguir o procedimento:

- Cálculo da área transversal total da coluna;

$$ A_t = \frac{(\pi \cdot D_c^2)}{4}$$

- Cálculo da área de escoamento;

$$ A_e = \epsilon \cdot A_t $$

- Cálculo do fluxo de gás;

$$ \phi = \frac{vaz_m}{A_e} $$

- Cálculo do $Y_{cgcp}$
- Determinação do $X_{cgcp}$
- Cálculo do fluxo líquido máximo;
- Cálculo da vazão mássica máxima do solvente.

```python
# Area transversal:
a_T = (np.pi*(Dc**2))/4
a_T
```

```python
# Area de escoamento:
a_E = e * a_T
a_E #m²
```

```python
# Fluxo do gás:
phi_gas = F_mass_segundo/a_E
phi_gas #kg/m²*s
```

```python
# Calculo de Y_cgcp
Y_cgcp = ((phi_gas**2)*Cf*(psi)*(mi_agua**0.2))/(rho_gas*rho_agua*C)
Y_cgcp
```

Retirando $X_{cgcp}$ do gráfico, para a linha de inundação, temos:

```python
X_cgcp = 1.6
```

```python
# Cálculo do fluxo líquido máximo:
phi_l_max = X_cgcp * phi_gas * ((rho_agua/rho_gas)**0.5)
phi_l_max #kg/m²*s
```

```python
# Cálculo da vazão mássica máxima de solvente:
m_l_max = phi_l_max * a_E
m_l_max #kg/s
```

Cálculo da área molhada:



```python
def f_onda_et_al_1968_area_molhada (a,sigmaC,sigmaL,ul,vl,g,rhol):
  #Eq.11.23
  # Encontrada na pág 642- livro Fundamentos de transferência de massa-Cremasco
  ReL = (ul/(a*vl))
  FrL = (a*ul*ul)/g
  weL = (rhol*ul*ul)/(a*sigmaL)
  aw = (1 - np.exp(-1.45*((sigmaC/sigmaL)**(0.75))*(ReL**(0.1))*(FrL**(-0.05))*(weL**(0.2))))*a
  #as:Area superficial do Recheio Seco 
  #Re:Numero de Reeynolds
  #Sc:Numero de Schimidt  
  #vL; viscosidade cinemática (m²/s)
  #M;massa molar
  #u;Velocidade superficil(m/s)
  return(aw)
```

```python
aw = f_onda_et_al_1968_area_molhada(a,sigmaC,sigmaL,ul,vl,g,rhol)
aw
```

Para a fase gasosa:

```python
DABG = dif #m²/s
DABl = 2.07e-9  # m²/s difusividade de SO_2 em água está dissolvido
ds = 10e-3 # diâmetro nominal do rechio em m
uG = u_g #m/s
# viscosidade cinemática
vG = 1.33e-5 #m²/s - do ar
```

```python
def f_onda_et_al_1968_fase_gasosa(a_AS,DABG,ds,P,R,Temp,uG,vG):
  #Eq.11.23
  # Encontrada na pág 642- livro Fundamentos de transferência de massa-Cremasco
  ReG = uG/(a_AS*vG)
  ScG = vG/DABG
  ky = 5.23*a_AS*DABG*((ds*a_AS)**(-2))*(P/(R*Temp))*(ReG**(0.7))*(ScG**(1/3))
  #as:Area superficial do Recheio Seco 
  #DAB:Coef de Difusividade 
  #P:Pressão 
  #ReG:Numero de Reeynolds(Gas)
  #Sc:Numero de Schimidt  (Gas)
  #vL; viscosidade cinemática (m²/s)
  #ds:dimetro nominal 
  #M;massa molar
  #u;Velocidade superficil(m/s)
  return(ky)
```

```python
P_Pa = 101325 # Pa
R = 8.314 * 1000  # Pa * m3 / (kmol * K)
```

Para a fase líquida:


















```python
# Cálculo da fase líquida
def f_onda_et_al_1968_fase_liquida(aw,a,DABl,ds,rhol,vl,g,uL,Ml):
  #eq.11.24
  # Encontrada na pág 642- livro Fundamentos de transferência de massa-Cremasco
  ReL = uL/(a*vl)
  ScL = vl/DABl
  kx = (5.1*0.001)*(rhol/Ml)*((g*vl)**(1/3))*((ds*a)**(0.4))*(((a/aw)*(ReL))**(2/3))*(ScL**(-0.5))
  #a:Area superficial do Recheio Seco 
  #DAB:Coef de Difusividade de A em B 
  #P:Pressão 
  #ReG:Numero de Reeynolds(Liquido)
  #Sc:Numero de Schimi#vL; viscosidade cinemática (m²/s)
  #ds:dimetro nominal 
  #M;massa molar 
  #u;Velocidade superficil(m/s)
  #rhol ; massa específica
  #DICA VIDE O EXEMPLO 11.3 PROPRIEDADES DA H2O
  return(kx)
```

```python
kx = f_onda_et_al_1968_fase_liquida(aw,a,DABl,ds,rho_agua,vl,g,u_l,MM_agua)
kx
```
