## Parte 1: Estudo da vazão mínima de solvente necessária para tratar 1000 kg/h da mistura gasosa.

Para isso podemos usar a seguinte equação: 

$$L_S^{min} = G_I \cdot \frac{(Y_F^A - Y_T^A)}{( X^{A*} - X_T^A)}$$


Ao usarmos essa equação precisaremos do gráfico Y_a x X_a para nos auxiliar.



```python
# Preparando kernel
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Carregando os dados iniciais: 

yaf = 0.25
Xat = 0
Xaf = 0.04
```

Gerando o gráfico Y_a x X_a:

Definindo 100 mol de mistura inicial de ar + acetona, temos que:

Como 25% é acetona, temos 75 mol de ar e 25 mol de acetona

Como 10% de acetona sobra no final do processo, teremos 25 mol * 10% = 2.5 mol de acetona

Logo,

**yat = 2.5/(75 + 2.5)**

```python
yat = 2.5/(75 + 2.5)
yat
```

```python
# Valores de y no equilíbrio (topo e fundo):

ya_eq = np.linspace(yat,yaf,20)
ya_eq
```

```python
# Calculando Y_a:

Y_a = ya_eq/(1 - ya_eq)
Y_a
```

```python
# calculando x no equilíbrio: 

xa_eq = ya_eq/1.9
xa_eq
```

```python
# Calculando X_a:

X_a = xa_eq/(1 - xa_eq)
X_a
```

```python
# Plotando a reta de equilíbrio:

fig1, ax1 = plt.subplots(figsize =(8,8))
ax1.plot(X_a,Y_a,label='Equilíbrio')
# Adicionando texto nos eixos - descrição
ax1.set_xlabel('X_a')
ax1.set_ylabel('Y_a')
ax1.grid()
ax1.legend()
```

Gerando a reta de operação:

Sendo que:

 - Yat = 0.03

 - Yaf = 0.33

 - Xat -> Valor decorrente de xa=0

 - Xaf -> Valor decorrente de xa=0.04

```python
# Calculando Xat e Xaf

Xat = 0/(1+0)
Xaf = 0.04/(1-0.04)
```

Agora podemos calcular também a vazão molar da mistura gasosa que corresponde somente ao ar (ou seja, ao componente inerte).

Lembrando que a mistura é composta por 25% acetona e 75% ar.

```python
# Calculo da massa molar média da mistura gasosa:
MM_acetona = 58.8 # kg/kmol
MM_ar = 16*0.21 + 0.79*28  # 21% O2 e 79% N2
```

```python
MM_mist = 0.25*MM_acetona + 0.75*MM_ar
MM_mist # kg/km
```

```python
# Calculo da vazão molar do solvente:

G = 1000/MM_mist
G # 29.57 kmol/h
```

```python
# Calculo da vazão referente ao ar (75% molar):

GI = G*0.75
GI # kmol/h
```
